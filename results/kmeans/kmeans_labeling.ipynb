# kmeans clustering
assign 10 clusters based on ADT-antibody surface marker expression

```python
import os, sys
import numpy as np
import pandas as pd
```

```python
os.listdir("./filtered")
```

```python
data_adtclr = pd.read_csv("./filtered/CBMC_ADT_clr_filtered.csv")
```

```python
data_adtclr
```

```python
X = data_adtclr.iloc[:,1:].values
X
```

```python
Xtrans = np.transpose(X)
Xtrans.shape
```

```python
from sklearn.cluster import KMeans

kmeans_instance = KMeans(n_clusters=10, random_state=2017, n_jobs=8).fit(Xtrans)
```

```python
y = kmeans_instance.labels_
```

```python
#cluster counts (8005 total cells)
np.unique(y, return_counts=True)
```

```python
y.shape
```

```python
#1d array to 2d array (so savetxt will write properly)
y = np.reshape(y,(-1,1))
np.transpose(y)
```

```python
#8005 columns, 1 row
#10 clusters are integer values from 0 to 9
np.savetxt("kmeans_10clusters_ADTclr_filtered.csv", np.transpose(y), delimiter=',', fmt='%i')
```
