```python
from bokeh.io import output_notebook, show
from bokeh.models import ColumnDataSource
```

```python
import numpy as np
```

```python
import bokeh.plotting as bp
```

```python
output_notebook()
```

```python
tsne_coords_names = np.genfromtxt("tsne_human.txt", dtype=None,usecols=0)
```

```python
tsne_coords_x = np.genfromtxt("tsne_human.txt", dtype=None, usecols=1)
```

```python
tsne_coords_y = np.genfromtxt("tsne_human.txt", dtype=None, usecols=2)
```

```python
ADT_kmeans = np.genfromtxt("kmeans_10clusters_ADTclr_filtered.csv", delimiter=',',dtype="int")
```

```python
rna_kmeans = np.genfromtxt("kmeans_10clusters_rna_filtered.csv", delimiter=',',dtype="int")
```

```python
data=dict()
data["x"]=tsne_coords_x.tolist()
data["y"]=tsne_coords_y.tolist()
tsne_coords_names = tsne_coords_names.tolist()
data["adt_kmeans"] = ADT_kmeans.tolist()
data["rna_kmeans"] = rna_kmeans.tolist()

for idx, x in enumerate(tsne_coords_names):
    x = x.decode('utf-8')
    tsne_coords_names[idx] = x
    
data["desc"] = tsne_coords_names
```

```python
from bokeh.palettes import Category10,Spectral,Set3,all_palettes


color_list = np.asarray(all_palettes['Paired'][10])
#color_list


colors = color_list[data["adt_kmeans"]]

data["colors"] = colors.tolist()
```

```python
from bokeh.models import HoverTool
from numpy import *
source = ColumnDataSource(data=data)

hover = HoverTool(
        tooltips=[
            ("index", "$index"),
            ("coord", "($x, $y)"),
            ("name", "@desc"),
            ("rna_kmeans_clust", "@rna_kmeans"),
            ("adt_kmeans_clust", "@adt_kmeans")
        ]
    )

p = bp.figure(plot_width=900, plot_height=700, title="ADT t-SNE with ADT K-Means Colors",
       tools=[hover],
       x_axis_type=None, y_axis_type=None, min_border=1)

p.scatter(
    'x',
    'y',
    fill_color='colors',
    source=source,
)
show(p)
```
